# 1. Water tunnel test

a. Calculate the Reynolds number for an F-18 flying at $60$ m/s. Use air density $\rho = 1$ kg/m$^3$ and dynamic viscosity of air $\mu = 10^{-5}$ Pa$\cdot$s. The F-18’s length is L = 17 m.

b. Aeronautical engineers at the NASA Dryden Flight Research Center placed a 1:48 scaled model of an F-18 in a water tunnel to visualize the flow around the aircraft (see image below). If the engineers were to match the Reynolds number when the aircraft flies at $60$ m/s, what should be the uniform velocity of the water in the water tunnel’s test section? Use water density $\rho = 1000$ kg/m$^3$ and water dynamic viscosity $\mu = 10^{−3}$ Pa$\cdot$s.


Screen Shot 2021-11-28 at 6.58.28 PM.png

## Solution

$$Re = \frac{\rho d V}{\mu}$$

We will use the length of the F-18 as $d$, our length dimension since the flow is acting on the entire body of the aircraft. 

$$Re_{1:1} = \frac{1\cdot17\cdot60}{10^{-5}}=102\times10^6$$

To match the Reynolds number, we will consider

$$Re_{1:1} = Re_{1:48}=\frac{\rho_{H_2O}\left(\frac{d}{48}\right)V_{H_2O}}{\mu_{H_2O}}$$

Which becomes

$$V_{H_2O} = \frac{Re_{1:1}\mu_{H_2O}}{\rho_{H_2O}\left(\frac{d}{48}\right)} = 288 m/s$$

# Laminar pipe flow

Consider steady fully-developed laminar flow in a circular pipe. We have shown in class that the velocity distribution is

$$u_z(r) = \frac{1}{4\mu}\frac{\Delta p}{l}(R^2-r^2)$$

a. Find the expression for the shear stress $\tau$ as a function of radius $r$.

b. Find the expression for the shear at the pipe wall $\tau_w$.

c. Find the force on a pipe section of length $l$ due to friction with the fluid.

d. Use control volume analysis to find the force on a pipe section of length $l$. Does your result agree with (c)?

## Solution

### Part a.

Newtonian shear stress is given by

$$\tau = \mu \frac{\partial u}{\partial y}$$

In polar coordinates though

$$y = r sin(\theta)$$

But since we are only considering the shear stress as a function of $r$, we will hold $\theta$ constant, essentially only looking in one veritcal plane and therefore

$$dy=dr$$

To solve our equation, we will find

$$\frac{\partial u}{\partial r} = -\left(\frac{1}{2\mu}\frac{\Delta p}{l}\right)r$$

which means

$$\tau = -\frac{\Delta p}{2l}r$$

### Part b.

The shear at the pipe wall is when $r=R$ therefore

$$\tau_w = -\frac{\Delta p}{2l}R$$

and the negative just describes the direction of the shear.

### Part c.

The force on the pipe wall is

$$F = \tau_w A$$

and for a circular pipe with length $l$

$$A = 2\pi R l$$

Which means

$$F = -\pi R^2\Delta p$$

### Part d.

Momentum Conservation

$$\int_{cs}\vec{V}\rho\vec{V}\cdot\hat{n}ds=p_1A_c - p_2 A_c+F_w$$

where

$A_c = \pi R^2$

However since our velocity in the pipe doesn't change, our momentum never changes either so the left side of the equation equals zero

This means our equation becomes

$$-\pi R^2\Delta p = F_w$$

Which is the same as what we found in part c



# Moody diagram

Oil with density $\rho = 800$ kg/m$^3$ and viscosity $\mu = 1.5 × 10^{−3}$ Pa$\cdot$s flows in forged steel pipe of diameter $D=0.2$m. The flowrate is $Q=0.1$m$^3$/s. The end of the pipe is open to the atmosphere. If the pipe has length $l = 800$m, what is the pressure at the other end of the pipe? Use the Moody diagram of Figure 2.

Screen Shot 2021-11-28 at 8.00.05 PM.png

## Solution

From our given information, we know that Renolds number for our system is given by

$$Re = \frac{\rho V D}{\mu}= \frac{\rho \frac{Q}{\pi D^2/4} D}{\mu} = 0.34\times 10^{6}$$

And the normalized roughtness is 

$$\frac{\epsilon}{D} = \frac{0.025}{200} = 1.2\times 10^{-4}$$

Using these points on the Moody diagram, we find our friction factor is approximently 0.015.

We then know that friction factor is given by

$$f = \frac{p_1-p_2}{0.5\rho V^2}\cdot\frac{D}{l}$$

Knowing $p_2 = 0$, we can rearrange the equation to find $p_1$ 

$$p_1 = \frac{0.5f\rho V^2 l }{D} = 194.5 kPa$$
