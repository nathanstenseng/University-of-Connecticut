# 1 NASA’s “Forever” rocket

Finally, NASA appears to be on the verge of an important discovery. Your colleagues at Marshall came up with a revolutionary rocket design. Your colleagues claim that their rocket is capable of reaching the edge of the universe. Your group supervisor asked you to review their proposal (this actually happens very often).

The rocket is initially filled with $30$ kg of gas. The nozzle exit has area $A = 2$ m$^2$ and the gas at the nozzle exit has density $\rho = 1.5$ kg/m$^3$. The uniform velocity of the gas at the nozzle exit is a function of time $t$ and it is given by:

$$V(t) = \frac{10}{(t+1)^2}\quad m/s$$


a. Is the flow at the nozzle exit steady or unsteady?

b. What is the mass flow rate at the nozzle exit?

c. What is the mass of the gas remaining in the rocket as a function of time? 

d. How long will it take for the gas to run out?

Screen Shot 2021-10-14 at 11.06.41 PM.png

## Solution

### Part a.

We can determine if the flow is steady or unsteady by taking the partial derivitive of the velocity equation:

$$\frac{\partial}{\partial t}\left(\frac{10}{(t+1)^2}\right) = \left(\frac{-20}{(t^2+1)^3}\right)\ne0$$

The flow out of the nozzel is unsteady.

### Part b.

We know the mass flow rate out of the rocket is given by:

$$\dot{M}=\rho Q$$

Where $Q$ is the volumetric flow rate given by:

$$Q = VA$$

Plugging in the given values from our problem statement gives the mass flow rate as:

$$\dot{M}(t) = \frac{-30}{(t+1)^2} \quad \frac{kg}{s}$$

### Part c.

The mass left in the rocket is $M_o+\int_0^t\dot{M}(x)dx$ which comes from the fundamental theorem of calculuse

Plugging in values from the problem statement and evaluating the integral gives:

$$M_{remaining}(t) = 30 + \left(\frac{30}{x+1}\right)_{x=0}^{x=t}$$

$$M_{remaining}(t) = 30 + \left(\frac{30}{t+1}-30\right)$$

$$M_{remaining}(t) = \frac{30}{t+1}$$



### Part d.

We can say the gas is out when $M_{remaining}(t) = 0$:

$$0=\frac{30}{t+1}$$

This equation has no real solutions for $t$, it only apporaches 0 at $t=\infty$,and therefore according to the equations, the rocket will never run out of fuel, being a forever rocket

# 2 Control volume choice

We will compute the anchoring force of Example 5.10 of the textbook using a different shape of control volume. Please see the class notes for the control volume shape and results. Here, we will use the values for the velocity and area we used in class: $V_1 = 10$ m/s, $A_1 = A_2 = 2$ m$^2$, and density $\rho = 1000$ kg/m$^3$. We found in class that the force is:

$$\vec{F} = −\rho V_1^2A_1(1 − cos(\theta))\hat{i} + ρV_1^2A_1 sin(\theta)\hat{j}$$

Show that the anchoring force $\vec{F}$ is the same for both control volumes: the one we used in class and that of the one below. In your solution, clearly indicate were the two calculations differ.

Screen Shot 2021-10-14 at 11.12.32 PM.png

## Solution

Step 1: The control volume has already been defined for us.

Step 2: Define the normal vectors and velocity vectors

We will assume all the normal vectors are outward with the only important normal vectors being:

$$\hat{n}_1 = <-1,0,0>$$

$$\hat{n}_2 = <0,1,0>$$

We can than say that the vector forms of the velocities will be:

$$\vec{V}_1 = <V_1,0,0>$$

$$\vec{V}_2 = <V_2 cos(\theta), V_2 sin(\theta),0>$$

Step 3: Conservation of mass

Here, we can use the simplified conservation of mass equations:

$$V_1A_1=V_2A_2$$

This proves that $V_1=V_2$ since $A_1=A_2$ according to the problem statement.

Step 4: Conservation of moment

We know this equation is steady so we will ignore the unsteady term. We will also first analysis the x component of moment and then the y component:

$$X_{mom} = \int_{cs}u\rho\vec{V}\cdot\hat{n}ds$$

We can add the different surfaces to expand this equation to:

$$\int_{A_1}u_1\rho\vec{V}_1\cdot\hat{n}_1ds + \int_{A_{2'}}u_2\rho\vec{V}_2\cdot\hat{n}_2ds = F_x$$

Where $A_{2'}$ is the area of the pipe passing through the control volume, which via trigonomotry will be:

$$A_{2'} = \frac{A_2}{sin(\theta)}$$

Taking the inner product gives:

$$\vec{V}_1\cdot\hat{n}_1=<V_1,0,0>\cdot<-1,0,0> = -V_1$$

$$\vec{V}_2\cdot\hat{n}_2=<V_2cos(\theta),V_2sin(\theta),0>\cdot<0,1,0> = V_2sin(\theta)$$

We can now rearrange our momentum equation to read:

$$V_1\rho(-V_1)\int_{A_1}ds + V_2cos(\theta)\rho V_2sin(\theta)\int_{A_{2'}}ds = F_x$$

Since the rest of the equation is constant with respect to the surface, the integral simply becomes:

$$V_1\rho(-V_1)A_1 + V_2cos(\theta)\rho V_2sin(\theta)\frac{A_2}{sin(\theta)} = F_x$$

We can then recall that $V_1=V_2$ and $A_1=A_2$ to say:

$$F_x = -\rho V_1^2A_1 + \rho V_1^2A_1cos(\theta)$$

Which can be simplified to:

$$F_x = -\rho V_1^2A_1(1-cos(\theta))$$

We can next find the y moment:

$$Y_{mom} = \int_{cs}v\rho\vec{V}\cdot\hat{n}ds$$

We can add the different surfaces to expand this equation to:

$$\int_{A_1}v_1\rho\vec{V}_1\cdot\hat{n}_1ds + \int_{A_{2'}}v_2\rho\vec{V}_2\cdot\hat{n}_2ds = F_y$$

Accounting for the inner products and bring constants out to the front leaves us with:

$$V_2sin(\theta)\rho V_2sin(\theta)\int_{A_{2'}}ds=F_y$$

Solving our integral and substituting $V_1=V_2$ and $A_1=A_2$ gives:

$$F_y = \rho V_1^2A_1sin(\theta)$$

This means our anchoring force is given by:

$$\vec{F} = -\rho V_1^2A_1(1-cos(\theta))\hat{i} + \rho V_1^2A_1sin(\theta)\hat{j}$$

#3 No more submarines

It appears that this class does not like submarines. Thus, from now on, we will switch to airborne platforms. A model plane is placed in a circular wind tunnel of radius $r_0 = 0.5$ m. The wind tunnel test section is shown below. At the test section inflow, the velocity is uniform $V_1 = 10$ m/s and the gage pressure is $p_1 = 200$ Pa. At the test section outflow the velocity profile linearly increases with respect to the radius as shown below, $V_2(r) = \frac{V_or}{r_o}$ m/s, and the gage pressure is $p_2 = 100$ Pa.

Find:

a. The value of the maximum velocity $V_0$ at the test section outflow. 

b. The drag force on the model plane.

Assume steady incompressible flow and neglect the friction with the wind tunnel walls. Use $\rho = 1.2$ kg/m$^3$ for the density of air.

Helpful hits: You will need to work with cylindrical coordinates because the problem has cylindrical symmetry. At the wind tunnel exit the velocity depends on the coordinate r. Please be careful how you treat this in the integrals.

Also: this is a “real” problem. You can use this method to estimate the drag force on objects.

Screen Shot 2021-10-15 at 8.53.34 AM.png

## Solution

Part a.

We can use Bernouli's equation along the top streamline to say:

$$p_1 +\frac{1}{2}\rho V_1^2 = p_2 + \frac{1}{2}\rho V_2^2$$

Where on the top streamline, $V_2=V_o$:

$$V_o = \sqrt{\frac{p_1+\frac{1}{2}\rho V_1^2-p_2}{\frac{1}{2}\rho}}$$

```python
p_1 = Q_(200, "Pa")
p_2 = Q_(100, "Pa")
V_1 = Q_(10, "m/s")
rho = Q_(1.2, "kg/m^3")

V_o = np.sqrt((p_1 + 0.5*rho*V_1**2 - p_2)/(0.5*rho))
print(V_o.to("m/s").round(5))
```

To determine the drag force on the plane, we will say the $\sum F_x$ force that is changing the momentum of the fluid.

Step 1: We will say the control volume is the same as defined in the photo.

Step 2: The normal vectors in 3-D Cartesian coordinates $<x,y,z>$:

$$\vec{n}_1 = <-1,0,0>$$

$$\vec{n}_2 = <1,0,0>$$

For this problem though, we will convert into cylindrical coordinates of form by $<\theta,r,z>$ where the cylinder is along the tunnel which changes our normal vector to:

$$\hat{n}_1 = <0,0,-1>$$

$$\hat{n}_1 = <0,0,-1>$$

Step 3: The velocity vectors in 3-D Cartesian coordinates will be:

$$\hat{V}_1 = <V_1,0,0>$$

$$\hat{V}_2 = <\frac{V_or}{r_o},0,0>$$

In cylinderical forms these will be:

$$\vec{V}_1 = <0,0,V_1>$$

$$\vec{V}_2 = <0,0,\frac{V_or}{r_o}>$$

Step 4: Since we are in steady state, we can say conservation of mass tells us that:

$$\int_{cs} \rho\vec{V}\cdot\vec{n}dS=0$$

Since the boundries are both flat disk, we can split the function into:

$$\int_{S_1} \rho\vec{V}_1\cdot\vec{n}_1dS + \int_{S_2} \rho\vec{V}_2\cdot\vec{n}_2dS = 0$$

Plugging in our vectors gives:

$$\int_{S_1} \rho<0,0,V_1>\cdot<0,0,-1>dS + \int_{S_2} \rho<0,0,\frac{V_or}{r_o}>\cdot<0,0,1>dS = 0$$

Taking the inner products gives:

$$\int_{S_1} -V_1\rho dS + \int_{S_2} \rho\frac{V_or}{r_o}dS = 0$$

We can parametrize the surfaces as:

$$dS = rdrd\theta$$

Where $r$ is added because of the Jacobian in cylindrical coordinates. Which turns the intergrals into:

$$-V_1\rho\int_0^{2\pi}\int_{0}^{r_o}rdrd\theta+\frac{\rho V_o}{r_o}\int_0^{2\pi}\int_{0}^{r_o}r\cdot rdrd\theta$$

Solving this becomes:

$$-V_1\rho\pi r_o^2 + \frac{\rho V_o}{r_o}\frac{2\pi r_o^3}{3} = 0$$

Which can be rearranged to say:

$$V_o = \frac{3V_1}{2}$$

Step 5: Momentum conservation for our steady flow will be:

$$\int_{S_1} \vec{V}_1\rho\vec{V}_1\cdot\vec{n}_1dS + \int_{S_2} \vec{V}_2\rho\vec{V}_2\cdot\vec{n}_2dS = \sum \vec{F}$$

The only flow that is nonzero is in the cylindrical $z$ direction:

$$\int_{S_1} V_1 \rho(-V_1)dS + \int_{S_2} \frac{V_or}{r_o}\rho\frac{V_or}{r_o}dS = p_1A_1 - p_2A_2 + F_{drag}$$

We will use a similar substitution into cylindrical coordinates to get:

$$-V_1^2\rho\int_{0}^{2\pi}\int_{0}^{r_o}rdrd\theta + \frac{\rho V_o^2}{r_o^2}\int_{0}^{2\pi}\int_{0}^{r_o}r^2\cdot rdrd\theta$$

Evaluating this integral gives:

$$-V_1^2\rho\pi r_o^2 +  \frac{\rho V_o^2}{r_o^2}\frac{2\pi r_o^4}{4}=p_1\pi r_o^2 - p_2 \pi r_o^2 + F_{drag}$$

```python
p_1 = Q_(200, "Pa")
p_2 = Q_(100, "Pa")
V_1 = Q_(10, "m/s")
V_o = 3*V_1/2
rho = Q_(1.2, "kg/m^3")
r_o = Q_(0.5, "m")

F_drag = (-V_1**2 * rho * np.pi * r_o**2) + (rho * V_o**2 / r_o**2) * (2 * np.pi * r_o**4 / 4) - p_1 * np.pi * r_o**2 + p_2 * np.pi * r_o**2

print(F_drag.to("N").round(3))

```

The force of drag is calculated to be 66.76 Newtons to the left. This makes sense because it means momentum is being taken out of the right flowing air. Momentum taken out of the fluid means the plane is taking that momentum from the fluid which means the fluid is acting opposite to the plane.
