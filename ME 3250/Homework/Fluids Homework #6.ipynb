# 1. Example 3.1 revisted

In Example 3.1, the textbook claims that for a flow around a sphere of radius $a$, the velocity along the stagnation streamline is given by:

$$V(x)=V_o\left(1+\frac{a^3}{x^3}\right)\tag{1}$$

Please see the textbook for definition of the coordinate system and schematic. We will derive the velocity distribution using potential flow theory (we will assume steady, inviscid, and irrotational flow.) The flow around the sphere is the superposition of a uniform flow of velocity $=V_o\hat{i}$ and a three dimensional doublet with velocity potential:

$$\phi(x,y,z) = -K\frac{x}{\left(x^2+y^2+z^2\right)^\frac{3}{2}}\tag{2}$$

a. Determine the doublet strength $K$ for a sphere with radius $a$.

b. Show that the velocity on the stagnation steamline is given by $V(x)=V_o\left(1+\frac{a^3}{x^3}\right)$

c. Find the pressure $p(x)$ on the stagnation streamline using the Bernoulli equation. Use zero gage pressure at $x=-\infty$

d. Verify that $p(-a)$ is the stagnation pressure

## Solution

### Part a.

We know that the velocity potential for a sphere is given by:

$$\phi =\phi_{uniform}+\phi_{doublet}$$

Therefore in cartesian coordinates:

$$\phi =V_ox+\frac{-Kx}{(x^2+y^2+z^2)^{\frac{3}{2}}}$$

We also know that the stagnation line occurs along the x-axis, therefore $y=0$ and $z=0$. This means we should only have an x component to our velocity. Our velocity should then be given by:

$$\vec{V} = \frac{\partial \phi}{\partial x}\hat{i}+0\hat{j}+0\hat{k}$$

Taking the partial derivitives gives:

$$\frac{\partial \phi}{\partial x} = V_o + \frac{2K}{x^3}$$

We know from the problem statement what $V(x)$ is:

$$V_o + V_o\frac{a^3}{x^3} = V_o + \frac{2K}{x^3}$$

To make these statements equal: 

$$K = \frac{V_o}{2}a^3$$

### Part b.

We know that for our system:

$$\vec{V} = u\hat{i} + v\hat{j} + w\hat{k}$$

where along the x-axis:

$u = V_o + \frac{2K}{x^3}$

$v = 0$

$w = 0$

We also know from part a. that:

$K = \frac{V_o}{2}a^3$

Which means that the magnitude of our velocity will be:

$$V = \sqrt{u^2+v^2+w^2}$$

where plugging in our velocity components and our $K$ constant gives:


$$V = V_o + \frac{V_oa^3}{x^3}$$

### Part c.

Bernoulli's equation tells us:

$$p_o + \frac{1}{2}\rho V_o^2 + \rho g h_o = p(x) + \frac{1}{2}\rho V^2(x) + \rho g h(x)$$

Where the streamline runs along the x axis so everything is a function of only x and $h(x) = h_o = 0$. This rearranges our equation to:

$$p(x) = p_0 + \frac{1}{2}\rho V_o^2 - \frac{1}{2}\rho V^2(x)$$

Which are all known:

$$p(x) = p_0 + \frac{1}{2}\rho V_o^2 - \frac{1}{2}\rho V_o^2\left(1+\frac{a^3}{x^3}\right)^2$$

Foiling out the parathensis gives:

$$p(x) = p_0 + \frac{1}{2}\rho V_o^2 - \frac{1}{2}\rho V_o^2\left(1+\frac{2a^3}{x^3} + \frac{a^6}{x^6}\right)$$

This now simplifies to by plugging in $p_o$ and canceling out terms:

$$p(x) = \frac{1}{2}\rho V_o^2\left(\frac{-2a^3}{x^3} - \frac{a^6}{x^6}\right)$$



### Part d.

We will plug $x=-a$ into our answer from part c. to get:

$$p(-a) = \frac{1}{2}\rho V_o^2\left(\frac{-2a^3}{-a^3} - \frac{a^6}{a^6}\right)$$

Which simplifies to:

$$p(-a) = \frac{1}{2}\rho V_o^2(2-1)$$

Which gives a stagnation pressure of $\frac{1}{2}\rho V_o^2$

## Solution

### Part a.

First we will fill out more terms in our diagram

$h = 40$ m

$h_w = 12$ m

$V = 67$ m/s

$|\Gamma| = 1500$ m$^2$/s

The steamline potentials for this problem are

$$\psi_{total}(x,y) = \psi_{unifrom}+\psi_{Free Vortex} + \psi_{MirroredVortex}$$

We know that

$$\psi_{unifrom} = Uy + \frac{\Gamma}{2\pi}ln(r) - \frac{\Gamma}{2\pi}ln(r)$$

We can determine the ground speed of the aircraft knowing that

$$\vec{V} = <\frac{\partial \psi}{\partial y},\frac{-\partial \psi}{\partial x}>$$

Before we differenciate, we will need to change our coordinates to cartesian

$$\psi_{unifrom} = Uy + \frac{\Gamma}{2\pi}ln\left(\sqrt{x^2+(y-h)^2}\right) - \frac{\Gamma}{2\pi}ln\left(\sqrt{x^2+(y+h)^2}\right)$$

We can use log rules to simplify this to

$$\psi_{unifrom} = Uy + \frac{\Gamma}{4\pi}ln\left(x^2+(y-h)^2\right) - \frac{\Gamma}{4\pi}ln\left(x^2+(y+h)^2\right)$$

Now we can find the function for velocity

$$\frac{\partial \psi}{\partial y} = U+\frac{\Gamma}{2\pi}\frac{1}{x^2+(y-h)^2}\cdot (y-h)-\frac{\Gamma}{2\pi}\frac{1}{x^2+(y+h)^2}\cdot (y+h)$$

$$\frac{-\partial \psi}{\partial x} = 0 - \frac{\Gamma}{2\pi}\frac{1}{x^2+(y-h)^2}\cdot (x)+\frac{\Gamma}{2\pi}\frac{1}{x^2+(y+h)^2}\cdot (x)$$

This means the velocity vector is

$$\vec{V} = <\frac{\partial \psi}{\partial y},\frac{-\partial \psi}{\partial x}>$$

To find the airspeed over the wing, we will find 

$$\sqrt{u^2(0,h_w)+v^2(0,h_w)}$$

```python
zero_m = Q_(0,"m")
V = np.sqrt(u(zero_m,h_w)**2 + v(zero_m,h_w)**2)
print(V.round(2))
```
