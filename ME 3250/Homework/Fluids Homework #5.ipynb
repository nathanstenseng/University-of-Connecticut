# 1. Taylor–Green vortex

One of the few exact analytical solutions of the equations of fluid motion was found by G. I. Taylor and A. E. Green in 1937. The general form of the Taylor–Green vortex flow is:

$$\vec{V}(t,x,y) = \left( \alpha e^{-2t} sin(x)cos(y)\right)\hat{i} + \left( \beta e^{-2t} cos(x)sin(y)\right) \hat{j} \tag{1}$$

a. Is the flow steady or unsteady?

b. If the flow is incompressible, find the relation between $\alpha$ and $\beta$.

c. Find the expression for the vorticity field $\vec{\omega}(t,x,y)$.

d. If the kinematic viscosity is $\nu = 1$, the density $\rho$, and $\alpha = 1$, find the expression for the pressure field $p(t,x,y)$.

## Solution

### Part a.

To determine is the flow is steady or unsteady, we will see if:

$$\frac{\partial \vec{V}}{\partial t}=\vec{0}$$

Evaluting this by components gives:

$$\frac{\partial}{\partial t} \left(\alpha e^{-2t} sin(x)cos(y)\right) = -2\alpha e^{-2t} sin(x)cos(y)\neq 0$$

$$\frac{\partial}{\partial t} \left(\beta e^{-2t} cos(x)sin(y)\right) = -2 \beta e^{-2t} cos(x)sin(y) \neq 0$$

Since the time derivatives of the components are nonzero then our flow is unsteady.

### Part b.

If the flow is incompressible, then we can say:

$$\vec{\nabla}\cdot\vec{V} = 0$$

In cartesian coordinates this can be written as:

$$\frac{\partial u}{\partial x} + \frac{\partial v}{\partial y} = 0$$

We will first evaluate each derivative:

$$\frac{\partial u}{\partial x} = \alpha e^{-2t}cos(x)cos(y)$$

$$\frac{\partial v}{\partial y} = \beta e^{-2t}cos(x)cos(y)$$

Plugging these into our equation gives:

$$\alpha e^{-2t}cos(x)cos(y) + \beta e^{-2t}cos(x)cos(y) = 0$$

Which simplifies to: 

$$\alpha = -\beta$$

### Part c.

We know that the rotation for two dimensional fluid motion is given by:

$$\vec{\omega} = \frac{1}{2}(\frac{\partial v}{\partial x} - \frac{\partial u}{\partial y})$$

Each partial derivative is:

$$\frac{\partial v}{\partial x} = -\beta e^{-2t}sin(x)sin(y)$$

$$\frac{\partial u}{\partial y} = -\alpha e^{-2t}sin(x)sin(y)$$

Plugging this in means:

$$\vec{\omega}(t,x,y) = \left(\frac{-\beta + \alpha}{2}\right)e^{-2t}sin(x)sin(y)$$

### Part d.

Using Navier Stokes, we can say that:

$$\frac{\partial \vec{V}}{\partial t} + (\vec{V}\cdot\vec{\nabla})(\vec{V}) = \frac{-1}{\rho} \vec{\nabla} p + \nu\vec{\nabla}^2\vec{V}$$

Expanding this by components gives:

$$\hat{i}: \frac{\partial u}{dt} + u \frac{\partial u}{\partial x} + v \frac{\partial u}{\partial y} = \frac{-1}{\rho}\frac{\partial p}{\partial x} + \nu\left( \frac{\partial^2u}{\partial x^2} + \frac{\partial^2u}{\partial y^2} \right)$$

$$\hat{j}: \frac{\partial v}{dt} + u \frac{\partial v}{\partial x} + v \frac{\partial v}{\partial y} = \frac{-1}{\rho}\frac{\partial p}{\partial y} + \nu\left( \frac{\partial^2v}{\partial x^2} + \frac{\partial^2v}{\partial y^2} \right)$$

The following partial derivatives in the $\hat{i}$ component are:

$$u = e^{-2t}sin(x)cos(y)$$

$$\frac{\partial u}{\partial t} = -2u$$

$$\frac{\partial u}{\partial x} = e^{-2t}cos(x)cos(y)$$

$$\frac{\partial u}{\partial y} = -e^{-2t}sin(x)sin(y)$$

$$\frac{\partial^2 u}{\partial x^2} = -u$$

$$\frac{\partial^2 u}{\partial y^2} = -u$$

Plugging this into our $\hat{i}$ equation gives:

$$\frac{\partial p}{\partial t} = \rho \left[ -e^{-4t}cos^2(y)sin(x)cos(x) - e^{-4t}sin^2(y)sin(x)coos(x)\right]$$

We can simplifies using the pathagorean identity and integrate to get:

$$p = -\rho e^{-4t}\int sin(x)cos(x)dx+c(y)$$

Evaluating this integral gives:

$$p = \frac{-\rho e^{-4t}}{2}sin^2(x) + c(y)$$

We can then take the partial derivative to get:

$$\frac{\partial p}{\partial y} = c'(y)$$

Comparing this to our $y$ momentum will let us find what $c'(y)$ is and therefore what $c(y)$ is. The partial derivatives for the $y$ momentum equation are:

$$\frac{\partial p}{\partial y} = \rho\left[ -e^{-4t}sin^2(x)sin(y)cos(y) - e^{-4t}cos^2(x)sin(y)cos(y) \right]$$

Setting this equal to our other equation gives:

$$c'(y) = -\rho e^{-4t}sin(y)cos(y)$$

Integrating with respect to $y$ yields:

$$c(y) = \frac{-\rho e^{-4t}}{2}sin^2(y)$$

Which gives a pressure equation:

$$p = \frac{-\rho e^{-4t}}{2}\left(sin^2(x) + sin^2(y)\right)$$

# 2. Vortex

Vortices as usually made up from an inner core and an outer part. For instance, hurricanes have a distinctive “eye” in the center. Consider two vortices with the azimuthal velocity distributions. The velocity is given in cylindrical coordinates and the velocity vector is $\vec{V} = u_r\hat{r} + u_{\theta} \hat{\theta} + u_z \hat{k}$.

Vortex A:

$$ u_{\theta}(r) = 
\begin{cases} 
      2r & r \le 1 \\
      \frac{2}{r} & r > 1
  \end{cases}
  \tag{2}
$$

Vortex B:

$$ u_{\theta}(r) = 
\begin{cases} 
      r & r \le 1 \\
      \frac{1}{r^2} & r > 1
  \end{cases}
  \tag{3}
$$

The other velocity components are zero everywhere: $u_r = 0$ and $u_z = 0$.

For each of the two vortices:

a) Show that the flow is incompressible.

b) Calculate the vorticity as a function of radius

## Solution

### Part a)

For incompressible flow, we know that:

$$\vec{\nabla}\cdot\vec{V}=\vec{0}$$

Where in cylindrical coordinates we can say $\vec{\nabla} = <\frac{1}{r}\frac{\partial}{\partial r}r, \frac{1}{r}\frac{\partial}{\partial \theta},\frac{\partial }{\partial z}>$:

$$\vec{\nabla}\cdot\vec{V} = \frac{1}{r}\frac{\partial(r u_r)}{\partial r} +  \frac{1}{r}\frac{\partial u_{\theta}}{\partial \theta} + \frac{\partial u_z}{\partial z} $$

#### Vortex A

Taking each partial derivative for vortex A is:

$$\frac{\partial(r u_r)}{\partial r} = 0$$

$$\frac{\partial u_{\theta}}{\partial \theta} = 0$$

$$\frac{\partial u_z}{\partial z} = 0$$

Plugging these into our equation gives:

$$\vec{\nabla}\cdot\vec{V}=0$$

Which is incompressable flow.

#### Vortex B

Taking each partial derivative for vortex B is:

$$\frac{\partial(r u_r)}{\partial r} = 0$$

$$\frac{\partial u_{\theta}}{\partial \theta} = 0$$

$$\frac{\partial u_z}{\partial z} = 0$$

Plugging these into our equation gives:

$$\vec{\nabla}\cdot\vec{V}=0$$

Which is incompressable flow.


### Part b)

We know that vorcitity is given by:

$$\vec{\zeta} = \vec{\nabla} \times \vec{V}$$

Again using our cylindrical del vector:

$$\vec{\nabla} = <\frac{1}{r}\frac{\partial}{\partial r}r, \frac{1}{r}\frac{\partial}{\partial \theta},\frac{\partial }{\partial z}>$$

and cylindrical velocity vector:

$$\vec{V} = <u_r,u_\theta,u_z>$$

This means that:

$$\vec{\zeta} = <\frac{1}{r}\frac{\partial u_z}{\partial \theta} - \frac{\partial u_\theta}{\partial z},\frac{\partial u_r}{\partial z} - \frac{\partial u_z}{\partial r},\frac{1}{r}\frac{\partial ru_\theta}{\partial r}-\frac{1}{r}\frac{\partial u_r}{\partial u_\theta}>$$

#### Vortex A:

We will only consider expressions involving $u_\theta$ since that is our only non-zero component. Infact, this only has 1 non-zero partial derivative:

$$\frac{\partial ru_\theta}{dr} = 
\begin{cases} 
      4r & r \le 1 \\
      2 & r > 1
  \end{cases}
$$

This means our vorticity will simplify to:

$$\vec{\zeta}(r) = <0,0,\begin{cases} 
      4 & r \le 1 \\
      \frac{2}{r} & r > 1
  \end{cases}>$$

  #### Vortex B:

We will only consider expressions involving $u_\theta$ since that is our only non-zero component. Infact, this only has 1 non-zero partial derivative:

$$\frac{\partial ru_\theta}{dr} = 
\begin{cases} 
      2r & r \le 1 \\
      \frac{-1}{r^2} & r > 1
  \end{cases}
$$

This means our vorticity will simplify to:

$$\vec{\zeta}(r) = <0,0,\begin{cases} 
      2 & r \le 1 \\
      \frac{-1}{r^3} & r > 1
  \end{cases}>$$

# 3. Divergence

The $x$-component of the velocity field in a two-dimensional incompressible flow is:

$$u(x,y) = sin(2x)sin(4y) \tag{4}$$

What is the $y$-component $v(x,y)$ of the velocity field?

## Solution

Since the flow is incompresible, we can say that:

$$\vec{\nabla}\cdot\vec V = 0$$

Which is equivilent to:

$$\frac{\partial u}{\partial x} + \frac{\partial v}{\partial y} = 0$$

Therefore:

$$-\frac{\partial u}{\partial x} = \frac{\partial v}{\partial y}$$

We can take the partial derivative to find that:

$$\frac{\partial u}{\partial x} = 2cos(2x)sin(4y)$$

and then we know:

$$v = \int -2cos(2x)sin(4y) dy + c(x)$$

Evaluating this integral gives:

$$v = \frac{1}{4}cos(2x)cos(4y) + c(x)$$

Where $c(x)$ is any function of $x$ since that would disapear when taking the $y$ partial derivative.
