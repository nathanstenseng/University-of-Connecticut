## Problem 3.7

Molybdenum (Mo) has a BCC crystal structure, an atomic radius of 0.1363 nm, and anatomic weight of 95.94 g/mol. Compute and compare its theoretical density with theexperimental value found inside the front cover of the book.

## Solution

Important Equations:

1. $\rho = \frac{nA}{V_C N_A}$

2. $a_{BCC}=\frac{4R}{\sqrt3}$

3. $V_C = a^3$

```python
R = Q_(0.1363, "nm")
A = Q_(95.94, "g/mol")
a = (4*R)/(np.sqrt(3))
V_C = a**3
n = 2
rho = (n*A)/(V_C*N_A)
print(rho.to("g/cm^3"))
```

The answer inside the back of the book is $10.22 g/cm^3$ and we calculated $10.2165 g/cm^3$

## Problem 3.9

Calculate the radius of a palladium (Pd) atom, given that Pd has an FCC crystal structure, a density of $12.0 g/cm^3$, and an atomic weight of $106.4 g/mol$.

## Solution

Important Equations:

1. $\rho = \frac{nA}{V_C N_A}\Rightarrow V_C = \frac{nA}{\rho N_A}$

2. $V_C = 16R^3\sqrt2 \Rightarrow R = (\frac{V_C}{16\sqrt2})^{\frac{1}{3}}$

```python
n = 4
rho = Q_(12.0, "g/cm^3")
A = Q_(106.4, "g/mol")

V_C = (4*A)/(rho*N_A)

R = ((V_C)/(16 * np.sqrt(2)))**(1/3)

print(R.to("nm"))
```

Palladium's atomic radius: $0.13756 nm$
