## Problem 12.3 

Show that the minimum cation-to-anion radius ratio for a coordination number of 6 is 0.414. [Hint:
Use the NaCl crystal structure (Figure 12.2), and assume that anions and cations are just touching along
cube edges and across face diagonals.]

## Solution

Proof: $\frac{r_C}{r_A} = 0.414$ for this case

For a corrdination number of 6, the atoms are arranged as shown in the image above. The closest two anions can get to eachother is $\overline{GF}$. 

Lets claim that $r_a+r_c = l$. This would imply that $\overline{GH} = l$, $\overline{HF} = l$, and $\overline{GF} = l\sqrt2$

But as stated before $\overline{GF} = 2r_a$

This yeilds $2r_a = r_a\sqrt2 + r_c\sqrt2$. Rearranging gets $\frac{r_C}{r_A} = \frac{2-\sqrt2}{\sqrt2} = 0.414$

## Problem 12.16

Calculate the theoretical density of NiO, given that it has the rock salt crystal structure.

## Solution

$$\rho = \frac{n'(\sum A_C + \sum A_A)}{V_C N_A}$$

Where: 

$n' = $ the number of units within a unit cell

$A_C = $ atomic weight of cations in the formula

$A_A = $ atomic weight of anions in the formula

$V_C = $ unit cell volume

$N_A = $ Avogadro's number

From table 12.4 we see that a rock salt crystal structure has an Anion Packing of FCC. For FCC, there are 4 atoms per unit cell. The atomic weight of Ni$^{2+}$, the cation, is 58.69 g/mol and the atomic weight of the anion O$^{2-}$ is 16.00 g/mol. The volume in the FCC is $a^3$ where $a$, the side length, is $r_A + 2r_C + r_A$. The radius of the ions are taken from table 12.3.

```python
n_prime = 4
A_C = Q_(58.69, "g/mol")
A_A = Q_(16, "g/mol")
r_C = Q_(0.069, "nm")
r_A = Q_(0.14, "nm")
a = r_A + 2*r_C + r_A
V_C = a**3
rho = (n_prime*(A_C + A_A))/(V_C*N_A)
print(round(rho.to("g/cm^3"),3))
```

## Problem 12.17 

Iron oxide (FeO) has the rock salt crystal structure and a density of 5.70 g/cm$^3$.

(a) Determine the unit cell edge length.

(b) How does this result compare with the edge length as determined from the radii in Table 12.3, assuming that the Fe$^2+$ and O$^2–$ ions just touch each other along the edges?

## Solution

$$(\rho = \frac{n'(\sum A_C + \sum A_A)}{V_C N_A}) \Rightarrow (V_C = \frac{n'(\sum A_C + \sum A_A)}{\rho N_A})$$

Where: 

$n' = $ the number of units within a unit cell

$A_C = $ atomic weight of cations in the formula

$A_A = $ atomic weight of anions in the formula

$V_C = $ unit cell volume

$N_A = $ Avogadro's number

From table 12.4 we see that a rock salt crystal structure has an Anion Packing of FCC. For FCC, there are 4 atoms per unit cell. The atomic weight of Fe$^{2+}$, the cation, is 55.85 g/mol and the atomic weight of the anion O$^{2-}$ is 16.00 g/mol. The volume in the FCC is $a^3$ where $a$, the side length.

```python
rho = Q_(5.70, "g/cm^3")
n_prime = 4
A_C = Q_(55.85, "g/mol")
A_A = Q_(16, "g/mol")

V_C = (4*(A_C+A_A))/(rho * N_A)
a = V_C**(1/3)
print(a.to("nm").round(4))
```
