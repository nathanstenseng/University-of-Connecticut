# Problem 9.58

(a) What is the distinction between hypoeutectoid and hypereutectoid steels?

(b) In a hypoeutectoid steel, both eutectoid and proeutectoid ferrite exist. Explain the difference between
them. What will be the carbon concentration in each?

# Solution

## Part (a)

Hypoeutectoid steel is steel that has a wt% to the left of the eutectoid point and hypereutectoid is to the right of the eutectoid point. 

## Part (b)

For a hypoeutectoid steel, the proeutectoid ferrite is a microconstituent that formed above the eutectoid temperature. The eutectoid ferrite is one of the constituents of pearlite that formed at a temperature below the eutectoid. The carbon concentration for both ferrites is 0.022 wt% C

# Problem 9.61

Consider 3.5 kg of austenite containing 0.95 wt% C and cooled to below 727C (1341F). 

(a) What is the proeutectoid phase?

(b) How many kilograms each of total ferrite and cementite form?

(c) How many kilograms each of pearlite and the proeutectoid cementite form? 

(d) Schematically sketch and label the resulting microstructure.



# Solution

## Part (a)

The proeutectoid phase is the phase with a temperature above the eutectoid point. The eutectoid point (going from one solid phase to two solid phases), occurs at 0.76 wt% Carbon and about 710 $^o$C. Our substance is to the right of the eutectoid point so we will have eith $\gamma$ phase (Austenite) or $\gamma$+$Fe_3C$ (Austenite and Cementite). For our temperature and composition, we will have a proeutectoid phase of $\gamma$+$Fe_3C$.

## Part (b)

Applying the lever rule, we can say

$$W_{\alpha} = \frac{C_{Fe_3C}-C_o}{C_{Fe_3C}-C_{\alpha}}=\frac{6.70-0.95}{6.70-0.022}=0.86$$

$$W_{Fe_3C} = \frac{C_o-C_{\alpha}}{C_{Fe_3C}-C_{\alpha}}=\frac{0.95-0.022}{6.70-0.022}=0.14$$

Multiplying these by our mass gives:

$$m_{\alpha} = 3.01 kg$$

$$m_{Fe_3C} = 0.49 kg$$

## Part (c)

In part (a), we determined that our substance is to the right of the eutectoid point. This means it is hypereutectoid. For hypereutectic steel, the following equations apply:

$$W_{pearlite} = \frac{6.70-C_o'}{5.97} = 0.96$$

$$W_{Fe_3C'} = \frac{C_o'-0.76}{5.97} = 0.03$$

Multiplying these by our mass gives:

$$W_{pearlite} = 3.36 kg$$

$$W_{Fe_3C'} = 0.105 kg$$

# Problem 9.67

The microstructure of an ironcarbon alloy consists of proeutectoid ferrite and pearlite; the mass fractions of these two microconstituents are 0.174 and 0.826, respectively. Determine the concentration of carbon in this alloy.

# Solution

If there is proeutectoid ferrite, then we must be to the left of the eutectoid point, because that is the only way to get ferrite above the eutectoid point. Now that we know this is hypoeutectoid, we know that:

$$W_{\alpha'}=\frac{0.76-C_o'}{0.74}=0.174$$

Solving for $C_o'$ means we have a carbon consentration of 0.63 wt% Carbon

# Problem 9.69

of these microconstituents are 0.11 and 0.89, respectively. Determine the concentration of carbon in this alloy.

# Solution

If there is cementite ($Fe_3C$), then we must be to the right of the eutectoid point while proeutectoid. Knowing we are hypereutectoid now, we can use the hypereutectoid equations:

$$W_{Fe_3C'}=\frac{C_o'-0.76}{5.97}=0.11$$

Solving for $C_o'$ means we have a carbon consentration of 1.42 wt% Carbon
