# Problem 9.21

A 65 wt% Ni–35 wt% Cu alloy is heated to a temperature within the α + liquid-phase region. If the composition of the α phase is 70 wt% Ni, determine:

(a) The temperature of the alloy

(b) The composition of the liquid phase 

(c) The mass fractions of both phases



# Solution

With a 70 wt% Ni $\alpha$ phase, we must be at 1350$^o$C. 

We will use the lever rule to determine or composition saying:

$C_{\alpha} = 70$ wt% Ni

$C_{o} = 65$ wt% Ni

$C_{l} \approx 59$ wt% Ni

According to lever rule:

$$W_{l} = \frac{C_{\alpha}-C_o}{C_{\alpha}-C_l}$$

$$W_{\alpha} = \frac{C_o-C_l}{C_{\alpha}-C_l}$$

Plugging values in yeilds

$W_l = 0.45$

$W_{\alpha} = 0.55$
