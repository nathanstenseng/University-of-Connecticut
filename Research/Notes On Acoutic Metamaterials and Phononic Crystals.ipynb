## Chapter 2: Discrete One-Dimensional Phononic and Resonant Crystals



### 2.1: One-Dimenaional Monoatomic Harmonic Crystal

**Defination**: An infinite chain of masses, $m$, with nearest neighbor interation modeled as a spring with constant, $\beta$. The seperation distance at rest between masses is, $a$. Atoms are indexed with, $n$, and have a displacement, $u$, with respect to their rest position.

**Governing equation**: without external forces, the equation of motion for atom $n$ is:

$$m\frac{d^2u_n}{dt^2}=\beta(u_{n+1}-u_n)-\beta(u_n-u_{n-1})\tag{2.1}$$

This will simplify to the form:

$$m\frac{d^2u_n}{dt^2}=\beta(u_{n+1}-2u_n+u_{n-1})\tag{2.2}$$

#### 2.1.1: Propagating waves

We will assume the solution to *Equation 2.2* is a propagating wave of the form:

$$u=Ae^{ikna}e^{i\omega t}\tag{2.3}$$

Where:

$n$ is the atom index

$a$ is the distance between atoms

$k$ is a wave number

$\omega$ is an angular frequency

Plugging in and solving gives the **Dispersion relation** for waves propagating in the medium:

$$\omega(k)=\omega_o\left|sin\left(k\frac{a}{2}\right)\right|\tag{2.5}$$

Where:

$\omega_o=2\sqrt{\frac{\beta}{m}}$ represnts the upper limit for angular frequency

Due to the repeating structer of the media, waves with length, $\lambda=\frac{2\pi}{k}$, which are larger than, $2a$, are physically equivalent to those smaller than $2a$, therefore we only need to graph the dispersion $\omega$ versus $k$ between $k \in \left[\frac{-\pi}{a},\frac{\pi}{a}\right]$.

This interval is the first **Brillouin zone** of the monoatomic periodic crystal medium we are working with.

#### 2.1.2: Phase and Group Velocity

The velocity the phase of the wave with spacific angular frequency, $\omega$, and specific wave vector, $k$, moves through the media is:

$$v_{\varphi}=\frac{\omega}{k}\tag{2.6}$$

*Note: Angular frequency, $\omega$, is cycles per time while wave number, $k$, is cycles per distance.*

The **group velocity** is the velocity of a wave packet (superposition of waves with a range of wave numbers over some interval)

Consider the superposition of waves 1 and 2 with properties $\omega_1 = \omega-\frac{\Delta \omega}{2}$, $\omega_2 = \omega+\frac{\Delta \omega}{2}$ and $k_1 = k-\frac{\Delta k}{2}$, $k_2 = k+\frac{\Delta k}{2}$

Assuming they have the same amplitude, these waves have the **displacement field** at mass $n$ of:

$$u^s_n=2Ae^{ikna}e^{i\omega t}cos\left(\frac{\Delta k}{2}na+\frac{\Delta\omega}{2}t\right)\tag{2.7}$$ 

This cosine term, known as a beat pulse, modulates the wave and travels with the group velocity:

$$v_g=\frac{\Delta \omega}{\Delta k}\tag{2.8}$$

For infinitesimally small differences in wave number and frequency, the group velocity is:

$$v_g=\frac{d\omega}{dk}\tag{2.9}$$

And recall that from *Equation 2.5*, frequency is a function of wave number, $\omega(k)$

For 1-D harmonic crystal, $v_g=\omega_o\frac{a}{2}cos\left(\frac{ka}{2}\right)$

**Energy sidenote**: *Proving that the velocity of the energy transported by the propagting wave is the same as the group velocity*

The average energy is the sume of potential and kinetic energies:

$$\langle E\rangle = \frac{1}{2}\beta(u_n-u_{n-1})(u_n-u_{n-1})^*+\frac{1}{2}m\dot{u}_n\dot{u}^*_n\tag{2.10}$$

Where $^*$ represents the complex conjugate

Inserting *Equation 2.3* and *Equation 2.5* into our average energy equation yields an average energy density:

$$\langle e\rangle = \frac{\langle E \rangle}{a}=4A^2\frac{\beta}{a}sin^2\left(k\frac{a}{2}\right)\tag{2.11}$$

The energy flow through one unit cell is  the real part of power, $\Phi$, defined as the product of the forces on mass $n$ due to one spring and the velocity of the mass:

$$\Phi = Re\{\beta(u_{n+1}-u_n)\dot{u}^*_n\}=\beta A^2 \omega_o\left|sin\left(k\frac{a}{2}\right)\right|sin(ka)\tag{2.12}$$

The velocity of the energy is the ratio of the energy flow, (2.12) to the average energy density (2.11) which is $v_e=\omega_o\frac{a}{2}cos(k\frac{a}{2})$. The same as the group velocity for a 1-D harmonic crystal.

#### 2.1.3: Evanescent Waves

Solving our equation of motion, *Equation 2.2*, for a nonpropagiting wave with an exponentially decaying amplitude is assumed to be of the form:

$$u_n=Ae^{-k''na}e^{ik'na}e^{i\omega t}\tag{2.13}$$

This comes about by assuming a complex wave number of the form $k=k'+ik''$.

Plugging this into the equaiton of motion and solving gives a **dispersion relation** of:

$$\omega = \omega_o cosh\left(k''\frac{a}{2}\right)\tag{2.15}$$


Earlier we solved the frequency as a function of wave number. Below we will also show an approach to solve for the wave number as a function of frequency. This approach allows us to determine isofrequency maps when dealing with 2-D and 3-D structured. 

Start by rewriting *Equation 2.4* to:

$$-m\omega^2=\beta(e^{ika}-2+e^{-ika})\tag{2.16}$$

We will define a new variable: $X=e^{ika}$. This means *Equation 2.16* becomes a quadratic in terms of $X$:

$$X^2+\left(\frac{m}{\beta}\omega^2-2\right)X+1=0\tag{2.17}$$

Which has two solutions:

$$X=\frac{1}{\omega_o^2}(\omega_o^2-2\omega^2)\pm \frac{2}{\omega_o^2}\sqrt{\omega^2(\omega^2-\omega_o^2)}\tag{2.18a}$$

This gives both real or complex solutions depending on values of the angular frequency. For $\omega \le \omega_o$, we have:

$$X=\frac{1}{\omega_o^2}(\omega_o^2-2\omega^2)\pm \frac{2i}{\omega_o^2}\sqrt{\omega^2(\omega_o^2-\omega^2)}\tag{2.18b}$$

We can also generalize the problem to a complex wave number of the form $k=k'+ik''$. In this cases, we get the form:

$$X=e^{-k''a}cos(k'a)+ie^{-k''a}sin(k'a)\tag{2.19}$$

When we consider $\omega>\omega_o$, *Equation 2.18a* remains real. The solution to this can also be written as the real portion of *Equaation 2.19*. We will denote this as:

$$h^{\pm}(\omega)=Re\{e^{-k''a}cos(k'a)+ie^{-k''a}sin(k'a)\} = e^{-k''a}cos(k'a)$$

If we assume that $k'=\frac{\pi}{a}$, we can obtain that:

$$k''^{\pm}(\omega)=-\frac{1}{a}ln(-h^{\pm}(\omega))\tag{2.20}$$

#### 2.1.4 Green's Function Approach

Consider harmonic solution with angular frequency $\omega$. We can recast *Equation 2.2* into the form:

$$\frac{1}{m}[\beta u_{n+1}+(m\omega^2-2\beta)u_n+\beta u_{u-1}]=0\tag{2.21}$$

We can rewrite this in matrix form when appying it to all masses in the 1-D monoatmomic crystal:

$$\overleftrightarrow{H_o}\vec{u}=\frac{1}{m}
\begin{bmatrix}
\ddots&\vdots&\vdots&\vdots&\vdots&\vdots&\vdots&\vdots&\vdots\\
\cdots&0&\beta&-\gamma&\beta&0&0&0&\cdots\\
\cdots&0&0&\beta&-\gamma&\beta&0&0&\cdots\\
\cdots&0&0&0&\beta&-\gamma&\beta&0&\cdots\\
\vdots&\vdots&\vdots&\vdots&\vdots&\vdots&\vdots&\vdots&\ddots
\end{bmatrix}
\begin{bmatrix}
\vdots\\
u_{n-1}\\
u_n\\
u_{n+1}\\
\vdots
\end{bmatrix}=\begin{bmatrix}
\vdots\\
0\\
0\\
0\\
\vdots
\end{bmatrix}\tag{2.22}$$

Where $\gamma=2\beta-m\omega^2$. The operator $\overleftrightarrow{H_o}$, is a more compact representation of the dynamic matrix in *Equation 2.22*, and $\vec{u}$ is the vector whose compoenets are the displacements of every mass. With this notation, **Green's Function** $\overleftrightarrow{G_o}$ is defined by:

$$\overleftrightarrow{H_o}\overleftrightarrow{G_o}=\overleftrightarrow{I}\tag{2.23}$$

The identity matrix, $\overleftrightarrow{I}$ written in component form is:

$$\sum_{n''} H_o(n,n'')G_o(n'',n')=\delta_{nn'}\tag{2.24}$$

where we use the Kroenecker symbol $\delta_{nn'}$ to represent the components of the identity matrix, that is 1 when $n=n'$ and 0 when $n\neq n'$

Since $\overleftrightarrow{H_o}$ is tridiagonal (harmonic interations are limited to the frist nearest neighbors), *Equation 2.23* becomes:

$$\frac{1}{m}[\beta G_o(n+1,n')-\gamma G_o(n,n')+\beta G_o(n-1,n
)]=\delta_{nn'}\tag{2.25}$$

A physical represention of Green's function $G_o(n,n')$ is the displacement of mass $n$ when a unit external force is applied at the site of mass $n''$. The solution of *Equation 2.25* is known and has the general form:

$$G_o(n,n') = \frac{m}{\beta}\frac{t^{|n-n'|+1}}{t^2-1}\tag{2.26}$$

The quantity, $t$, is determined by inserting this gernal solution into *Equation 2.25* and choosing $n=n'$. In this case we obtain the simple quadratic equation:

$$t^2-2\zeta t+1=0\tag{2.27}$$

Where 

$\zeta = \frac{\gamma}{2\beta} = 1-\frac{m\omega^2}{2\beta}=1-\frac{2\omega^2}{\omega_o^2}$

The resolution of the quadratic equation yields:

$$t = \left\{
\begin{array}{ll}
      \zeta-\sqrt{\zeta^2-1} & \zeta>1 \\
      \zeta+i\sqrt{1-\zeta^2} & -1\leq\zeta\leq 1 \\
      \zeta+\sqrt{\zeta^2-1} & \zeta<-1 \\
\end{array} 
\right.\tag{2.28}$$

We can recast the operator, $\overleftrightarrow{H_o}$ as $\overleftrightarrow{H_o'}-\omega^2\overleftrightarrow{I}$ where the operator, $H'_o$ depends only on the spring constant $\beta$. We can then rewrite *Equation 2.23* to say:

$$\overleftrightarrow{G_o}=\overleftrightarrow{I}\left(\overleftrightarrow{H'_o}-\omega^2\right)^{-1}\tag{2.29}$$

meaning when the poles (zeros fo the denominator) of Green's function are eigenvalues of the operator, $\overleftrightarrow{H'_o}$. According to *Equation 2.26*, the poles of Green's function for a 1-D system are given by:

$$t^2-1=0\tag{2.30}$$

This condition is met when $t=e^{ika}$

### 2.2 Periodic One-Dimensional Harmonic Crystals

#### 2.2.1 One-Dimenaional Monoatomic Crystal and Super-Cell Approach

**Super-period**, $R=Na$: a super-cell representation of the crystal. A section of individual periodically repeated cells. $N$ being the number of masses in the super cell and $a$ being their distance.

We will solve the equation of motion for a super-cell of mass, $l$. In *Equation 2.21* this becomes:

$$-m\omega^2 u_l=\beta(u_{l+1}-2u_l+u_{l-1})\tag{2.31}$$

We will now assume the displacement obeys **Block's Theorem**. The solution are the product of plane waves and a periodic function of the super-cell structure:

$$u_1(k)=e^{ikla}\tilde{u}_l(k)\tag{2.32}$$

The periodic function $\tilde{u}_l(k)$ satisfies the conditions $\tilde{u}_1(k)=\tilde{u}_{l+N}(k)$. The wave number, $k$, is limited to $\left[\frac{-\pi}{R},\frac{\pi}{R}\right]$. The function has the Fourier series:

$$\tilde{u}_l(k)=\sum_gu_g(k)e^{igla}\tag{2.33}$$

Where the reciprocal lattice vector of the periodic structure of super-cells is $g=\frac{2\pi}{Na}n$ with $n$ being an integer.

Plugging this in to *Equation 2.31* and solving for $\omega$ gives the relation:

$$\omega(k) = \omega_o \left|sin\left((k+g)\frac{a}{2}\right)\right|\tag{2.35}$$

Recall that $g$ consist of all integers $n$. However many times, different integers will produce the same graphs.


**Example**: super cell that is 2a long, containing 2 masses. $N=2$.

For this, the reciprocal space vectors, $g=\frac{\pi}{a}n$ and the equation becomes

$$\omega(k) = \omega_o \left|sin\left(\left(k+\frac{\pi}{a}n\right)\frac{a}{2}\right)\right|\tag{2.35}$$

This has two unique dispersion relations, all other $n$ values repeat the first two. The solution can either be graphed or obtained by translating the the wave number axis by $\frac{\pi}{a}$.

The super-cell representation can be obtained also graphically by folding the full dispersion curve about vertical lines. 

In general, one can construct the band structure by folding a single dispersion curve $N$ times inside a Brillouin zone: $\left[0,\frac{\pi}{Na}\right]$

#### 2.2.1 One-Dimensional Diatomic Harmonic Crystal

One-dimensional repeated patterns consisting of two atoms. For example, a pair of two masses, an even mass and an odd mass, that is repeated infinitly along the chain.

The equation of motion of of two adjacent odd and even atoms are:

$$\left\{
\begin{array}{ll}
      m_1\ddot{u}_{2n} = \beta(u_{2n+1}-u_{2n}+u_{2n-1})\\
      m_2\ddot{u}_{2n+1} = \beta(u_{2n+2}-u_{2n+1}-u_{2n})\\
\end{array} 
\right.\tag{2.36}$$

We want solutions for propagating waves with different amplitudes for each of the two types of atoms:

$$\left\{
\begin{array}{ll}
      u_{2n} = Ae^{i\omega t}e^{ik2na}\\
      u_{2n+1} = Be^{i\omega t}e^{ik(2n+1)a}\\
\end{array} 
\right.\tag{2.37}$$

Plugging these solution forms into *Equation 2.36* gives two linear equations:

$$\left\{
\begin{array}{ll}
      (2\beta-m_1\omega^2)A-2\beta cos(ka)B = 0\\
      -2\beta cos(ka)A+(2\beta-m_1\omega^2)B = 0\\
\end{array} 
\right.\tag{2.37}$$

This is an eigenvalue problem in $\omega^2$. To solve for non-trivial solutions ($A\ne 0, B\ne 0$), we will see when the determinant of the matrix is equal to zero. First, we will substitute $\alpha=\omega^2$. The determinate set equal to zero gives us:

$$\alpha^2-2\beta\left(\frac{1}{m_1}+\frac{1}{m_2}\right)\alpha + \frac{4\beta^2}{m_1m_2}sin^2(ka)=0\tag{2.40}$$

Which has two solutions:

$$\omega^2=\alpha=\beta\left(\frac{1}{m_1}+\frac{1}{m_2}\right)\pm\sqrt{\beta^2\left(\frac{1}{m_1}+\frac{1}{m_2}\right)^2-\frac{4\beta^2}{m_1m_2}sin^2(ka)}\tag{2.41}$$

These solutions are periodic in wave number, $k$ with a period of $\frac{\pi}{a}$, so we only need to look at that interval $\left[0,\frac{\pi}{a}\right]$. This interval is known as the irreducible Brillouin zone. 

Recall that *Equation 2.41* gives two functions for $\omega(k)$. Graphing both of those will give a band structure with two brances (Two since our system is diatomic). For two different masses, $m_1$ and $m_2$, the branches are seperated by a gap. The first branch ranges on the y-axis from $[0,\omega_1]$ and the second branch ranges from $[\omega_2,\omega_3]$ where: $\omega_1 = \sqrt{\frac{2\beta}{m_1}}, \omega_2=\sqrt{\frac{2\beta}{m_2}}, \omega_3 = \sqrt{2\beta\left(\frac{1}{m_1}+\frac{1}{m_2}\right)}$.

The **acoustic branch** is the low-frequency branch. The **optical branch** is the high-frequency branch. In the limit as the masses of the two atoms are equal, the diatomic crystal becomes a monoatomic crystal. 

It is imporatant to note that the acoustic branch has a positive slope but more importantly the optical branch has a negative slope. This suggest the group velocity and energy velocity points in the opposite direction of the wave vector. This is important when dealing with the concept of negative refraction. But it doesn't quite work in this one-dimensional case.
